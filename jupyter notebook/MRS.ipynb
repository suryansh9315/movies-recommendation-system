```python
import numpy as np
import pandas as pd
import ast
```

```python
movies_db = pd.read_csv("movies.csv")
credits_db = pd.read_csv("credits.csv")
```

```python
movies_db.info()
credits_db.info()
```

```python
movies_db.head(1)
```

```python
credits_db.head()
```

```python
movies = movies_db.merge(credits_db,on="title")
```

```python
movies.head(2)
```

```python
movies.info()
```

```python
# genres,id,keywords,title,overview,cast,crew
movies = movies[['id','genres','keywords','title','overview','cast','crew']]
```

```python
movies.head(1)
```

```python
movies.info()
```

```python
movies = movies.dropna()
movies.shape
```

```python
movies.duplicated().sum()
```

```python
movies['genres'][0]
```

```python
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(2)
```

```python
def convert5(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 5:
            l.append(i['name'])
            counter += 1
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert5)
```

```python
movies.head(1)
```

```python
def getDirector(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(getDirector)
```

```python
movies.head(3)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(3)
```

```python
def removeSpaces(obj):
    list = []
    for i in obj:
        list.append("".join(i.split()))
    return list
```

```python
movies['cast'] = movies['cast'].apply(removeSpaces)
```

```python
movies.head(2)
```

```python
movies['crew'] = movies['crew'].apply(removeSpaces)
movies['genres'] = movies['genres'].apply(removeSpaces)
movies['keywords'] = movies['keywords'].apply(removeSpaces)
```

```python
movies.head(2)
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] + movies['overview']
```

```python
movies.head(2)
```

```python
new_df = movies[['id','title','tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head(2)['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head(2)['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
 recommend('Batman Begins')
```
